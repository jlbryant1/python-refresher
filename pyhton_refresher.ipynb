```python
#List comprehensions
squares = [x**2 for x in range(10)] # Takes each number 0-9, squares it, and stores all results in a list
print (squares) # Prints Output to console
evens = [x for x in range(20) if x % 2 ==0] # Takes each number 0-19, but only keeps it if it's divisible by 2 (even)
print (evens) # Prints Output to console
```

```python
#Dictionary Comprehensions

# Loops through each word in the list and creates a dictionary
# where the key is the word and the value is its length
#VVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVV
word_lengths = {word:len(word)for word in ["James", "Lee", "Bryant", "III"]}
print (word_lengths) # Prints Output to console
```

```python
#Nested comprehensions
matrix = [[1,2,3], [4,5,6], [7,8,9]] # A list of lists — think of it as a 3x3 grid of numbers

# Loops through each row, then each value in that row,
# and pulls every value into one flat list
# "for row in matrix" grabs [1,2,3] then [4,5,6] then [7,8,9]
# "for val in row" grabs each individual number from that row
#VVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVV
flattened = [val for row in matrix for val in row]
print (flattened) # Prints Output to console
```

```python
#Default arguments, *args, *kwargs

# Defines a function with:
#   - "data" — a required argument (must be provided every time)
#   - "learning_rate" — optional, defaults to 0.01 if not provided
#   - "epochs" — optional, defaults to 10 if not provided
#   - "**kwargs" — catches any extra keyword arguments as a dictionary
#VVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVV
def train_model(data, learning_rate = 0.01, epochs=10, **kwargs):

    # f-string: the {} parts get replaced with the variable values
    #VVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVV
    print(f"Training with lr={learning_rate}, epochs={epochs}")

    # kwargs will be a dictionary of any extra arguments you passed in
    #VVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVV
    print (f"Extra params: {kwargs}")
    
train_model("my_data") # Call with just the required argument — defaults kick in

# Call with overrides and extra arguments — batch_size and verbose
# get captured by **kwargs since they're not named parameters
#VVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVVV
train_model("my_data", learning_rate=0.05, epochs=20,batch_size=32, verbose=True)
```

```python
# Imports the pandas library and gives it the shorthand name "pd"
# Pandas is the main library for working with tabular data in Python
import pandas as pd

# Create a CSV file
# A dictionary where each key is a column name
# and each value is a list of entries for that column
# None represents a missing value (will show as NaN in pandas)
data = {
    "name": ["Jay", "Breya", "Nigel", None],
    "age": [28, 23, 30, 22],
    "score": [85, 92, 78, 88]
    
}

# Converts the dictionary into a DataFrame — a table-like structure
# Think of it like a spreadsheet with rows and columns
df = pd.DataFrame(data)

# Saves the DataFrame to a CSV file called "data.csv"
# index=False prevents pandas from adding an extra numbered column on the left
df.to_csv("data.csv", index=False)
print("CSV created!")

# Defines a class that handles loading and cleaning data
class DataProcessor:

    # __init__ is the constructor — runs automatically when you create an instance
    # "self" refers to the specific instance being created
    def __init__(self, filepath):
        self.filepath = filepath
        self.data = None

    # Reads the CSV file into a pandas DataFrame and stores it in self.data
    def load(self):
        self.data = pd.read_csv(self.filepath)
        return self  # Returns the instance so you can chain methods like .load().clean()

    def clean (self):

        # dropna() removes any row that has a missing value (NaN/None)
        self.data = self.data.dropna()
        return self # Returns the instance for method chaining

# Creates a new DataProcessor instance, passing in the CSV file path
processor = DataProcessor("data.csv")

# Calls the load method, which reads "data.csv" into processor.data
processor.load()
print("Before cleaning:")

# Prints the full DataFrame — should show 4 rows, one with NaN for the name
print(processor.data)

# Calls the clean method, which removes rows with missing values
processor.clean()

# \n adds a blank line before this label for readability
print("\nAfter cleaning:")

# Prints the cleaned DataFrame — the row with NaN should be gone
print(processor.data)
```
